```python
#import library
import numpy as np #untuk melakukan operasi matematika dan numerik, seperti pengolahan array dan matriks.
import imageio #library untuk membaca dan menulis berbagai format file gambar
import matplotlib.pyplot as plt #untuk membuat visualisasi grafik, diagram, dan plot data
```

```python
#membaca gambar
img = imageio.imread("matahari.jpg")
```

```python
#mendapatkan resolusi dan type dari gambar
img_height = img.shape[0] #Baris kode ini akan mengambil nilai dimensi height dari citra dan menyimpannya pada variabel
img_width = img.shape[1] #Baris kode ini akan mengambil nilai dimensi width dari citra dan menyimpannya pada variabel
img_channel = img.shape[2] # Baris kode ini akan mengambil nilai dimensi channel dari citra dan menyimpannya pada variabel
```

```python
#membuat variabel img_inversi
img_inversi = np.zeros(img.shape, dtype=np.uint8) #Array ini nantinya dapat digunakan untuk menyimpan gambar hasil inversi.
```

```python
#Membuat fungsi untuk inversi grayscale
def inversi_grayscale(nilai):
    for y in range(0, img_height): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi height dari citra
        for x in range(0, img_width): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi width dari citra 
            red = img[y][x][0] # akan mengambil nilai piksel untuk warna merah (red) pada koordinat (y, x) 
            green = img[y][x][1] # akan mengambil nilai piksel untuk warna hijau (green) pada koordinat (y, x) 
            blue = img[y][x][2] # akan mengambil nilai piksel untuk warna biru (blue) pada koordinat (y, x) 
            gray = (int(red) + int(green) + int(blue)) / 3 #akan menghitung nilai rata-rata dari ketiga nilai piksel warna (red, green, dan blue) pada koordinat (y, x) 
            gray = nilai - gray
            img_inversi[y][x] = (gray, gray, gray) #akan menempatkan nilai piksel yang sudah dikonversi menjadi citra grayscale pada koordinat (y, x)
```

```python
#Membuat fungsi untuk inversi rgb
def inversi_rgb(nilai): #  Fungsi yang melakukan inversi RGB
    for y in range(0, img_height): # Loop untuk setiap baris gambar
        for x in range(0, img_width): # Loop untuk setiap kolom gambar
            red = img[y][x][0] # Nilai komponen merah pada pixel saat ini
            red = nilai - red # Menghitung nilai inversi komponen merah
            green = img[y][x][1] # Menghitung nilai inversi komponen merah
            green = nilai - green # Menghitung nilai inversi komponen hijau
            blue = img[y][x][2] # Nilai komponen biru pada pixel saat ini
            blue = nilai - blue # Menghitung nilai inversi komponen biru
            img_inversi[y][x] = (red, green, blue) # Set nilai pixel inversi pada gambar inversi
```

```python
#menampilkan hasil inversi
inversi_grayscale(255) # Untuk melakukan inversi gambar grayscale dengan nilai parameter 255
plt.imshow(img_inversi) # Menampilkan gambar hasil inversi
plt.title("Inversi Grayscale") # Memberikan judul pada gambar yang ditampilkan
plt.show() # Menampilkan gambar secara keseluruhan

inversi_rgb(255) # Untuk melakukan inversi gambar RGB dengan nilai parameter 255
plt.imshow(img_inversi) # Menampilkan gambar hasil inversi
plt.title("Inversi RGB") # Memberikan judul pada gambar yang ditampilkan
plt.show() # Menampilkan gambar secara keseluruhan
```

```python
#Membuat variabel img_log untuk menampung hasi
img_log = np.zeros(img.shape, dtype=np.uint8) # Menginisialisasi variabel img_grayscale dengan nilai zero array yang memiliki ukuran sama dengan i
```

```python
#Mendefinisikan fungsi untuk log
def log(c):    # mendefinisikan fungsi dengan nama 'log' dan parameter 'c'
    for y in range(0, img_height):    # melakukan perulangan untuk setiap baris gambar
        for x in range(0, img_width):    # melakukan perulangan untuk setiap piksel pada setiap baris
            red = img[y][x][0]    # mengambil nilai komponen merah piksel pada koordinat (x,y)
            green = img[y][x][1]    # mengambil nilai komponen hijau piksel pada koordinat (x,y)
            blue = img[y][x][2]    # mengambil nilai komponen biru piksel pada koordinat (x,y)
            gray = (int(red) + int(green) + int(blue)) / 3    # menghitung nilai rata-rata intensitas piksel untuk mendapatkan citra grayscale
            gray = int(c * np.log(gray + 1))    # mengaplikasikan fungsi logaritma pada nilai grayscale dengan parameter 'c'
            if gray > 255:    # memastikan bahwa nilai grayscale tidak melebihi 255 (nilai maksimum)
                gray = 255
            if gray < 0:    # memastikan bahwa nilai grayscale tidak kurang dari 0 (nilai minimum)
                gray = 0
            img_log[y][x] = (gray, gray, gray)    # menyimpan nilai grayscale hasil transformasi pada citra yang baru (img_log)
```

```python
#menampilkan hasil log
img_log = np.zeros(img.shape, dtype=np.uint8) # membuat matriks kosong dengan ukuran yang sama dengan gambar asli
log(30) # melakukan transformasi log pada gambar
plt.imshow(img_log) # menampilkan gambar hasil transformasi menggunakan matplotlib
plt.title("Log") # menambahkan judul pada plot gambar
plt.show() # menampilkan plot gambar
```

```python
#Membuat variabel img_inlog untuk menampung hasil
img_inlog = np.zeros(img.shape, dtype=np.uint8) # Membuat matriks kosong dengan ukuran yang sama dengan gambar asli
```

```python
#Mendefinisikan fungsi untuk inversi log
def inlog(c): # Membuat fungsi inlog dengan satu parameter input yaitu c
    for y in range(0, img_height): # Melakukan iterasi pada setiap pixel gambar dengan range dari 0 sampai tinggi gambar
        for x in range(0, img_width): # Melakukan iterasi pada setiap pixel gambar dengan range dari 0 sampai lebar gambar
            red = img[y][x][0] # Mengambil nilai komponen warna merah dari pixel gambar pada posisi (y, x)
            green = img[y][x][1] # Mengambil nilai komponen warna hijau dari pixel gambar pada posisi (y, x)
            blue = img[y][x][2] # Mengambil nilai komponen warna biru dari pixel gambar pada posisi (y, x)
            gray = (int(red) + int(green) + int(blue)) / 3 # Menghitung nilai rata-rata dari ketiga komponen warna pada pixel tersebut
            gray = int(c * np.log(255 - gray + 1)) # Menghitung nilai rata-rata dari ketiga komponen warna pada pixel tersebut
            if gray > 255: # Memastikan bahwa nilai grayscale tidak melebihi 255 (nilai maksimum)
                gray = 255
            if gray < 0: # Memastikan bahwa nilai grayscale tidak kurang dari 0 (nilai minimum)
                gray = 0
            img_inlog[y][x] = (gray, gray, gray) # Mengisi nilai pixel baru pada gambar output dengan nilai gray yang sudah dihitung
```

```python
#menampilkan hasil inversi log
inlog(30) # Untuk menghasilkan efek inversi dan log
plt.imshow(img_inlog) # Menampilkan citra hasil proses
plt.title("Inversi & Log") # Menambahkan judul
plt.show() # Menampilkan plot gambar
```

```python
#Membuat variabel img_nthpower untuk menampung hasil
img_nthpower = np.zeros(img.shape, dtype=np.uint8)
```

```python
#Mendefinisikan fungsi untuk nth power
def nthpower(c, y): # Mendefinisikan fungsi nthpower dengan dua parameter yaitu c dan y
    thc = c / 100 # Menghitung nilai threshold c (thc) dengan membagi input c dengan 100
    thy = y / 100 # Menghitung nilai threshold y (thy) dengan membagi input y dengan 100
    for y in range(0, img_height): # Melakukan iterasi pada setiap pixel baris gambar
        for x in range(0, img_width): # Melakukan iterasi pada setiap pixel kolom gambar
            red = img[y][x][0] # Menyimpan nilai warna merah dari pixel tertentu
            green = img[y][x][1] # Menyimpan nilai warna hijau dari pixel tertentu
            blue = img[y][x][2] # Menyimpan nilai warna biru dari pixel tertentu
            gray = (int(red) + int(green) + int(blue)) / 3 # Menghitung rata-rata nilai warna dari ketiga channel warna (merah, hijau, dan biru) untuk mendapatkan nilai grayscale
            gray = int(thc * pow(gray, thy)) # Menghitung nilai grayscale yang baru dengan memangkatkan nilai grayscale sebelumnya dengan threshold y, kemudian mengalikan dengan threshold c
            if gray > 255: # Memastikan bahwa nilai grayscale tidak melebihi 255 (nilai maksimum)
                gray = 255
            if gray < 0: # Memastikan bahwa nilai grayscale tidak kurang dari 0 (nilai minimum)
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray) # Menyimpan nilai grayscale yang baru ke dalam setiap channel warna dari pix
```

```python
#menampilkan hasil
nthpower(50, 100) # Memanggil fungsi nthpower dengan parameter
plt.imshow(img_nthpower) # Menampilkan citra hasil proses
plt.title("Nth Power") # Menambahkan judul
plt.show() # Menampilkan plot gambar
```

```python
#Membuat variabel img_nthrootpower
img_nthrootpower = np.zeros(img.shape, dtype=np.uint8) # Menginisialisasi variabel img_grayscale dengan nilai zero array yang memiliki ukuran
```

```python
#Membuat fungsi untuk nth root power
def nthrootpower(c, y): # Mendefinisikan fungsi nthrootpower dengan dua parameter yaitu c dan y
    thc = c / 100 # Menghitung nilai threshold c (thc) dengan membagi input c dengan 100
    thy = y / 100 # Menghitung nilai threshold y (thy) dengan membagi input y dengan 100
    for y in range(0, img_height): # # Melakukan iterasi pada setiap pixel baris gambar
        for x in range(0, img_width): # Melakukan iterasi pada setiap pixel kolom gambar
            red = img[y][x][0] # Menyimpan nilai warna merah dari pixel tertentu
            green = img[y][x][1] # Menyimpan nilai warna hijau dari pixel tertentu
            blue = img[y][x][2] # Menyimpan nilai warna biru dari pixel tertentu
            gray = (int(red) + int(green) + int(blue)) / 3 # Menghitung rata-rata nilai warna dari ketiga channel warna (merah, hijau, dan biru) untuk mendapatkan nilai grayscale
            gray = int(thc * pow(gray, 1./thy)) # Menghitung nilai grayscale yang baru dengan memangkatkan nilai grayscale sebelumnya ke pangkat 1/thy, kemudian dikalikan dengan thc
            if gray > 255: # Memastikan bahwa nilai grayscale tidak melebihi 255 (nilai maksimum)
                gray = 255
            if gray < 0: # Memastikan bahwa nilai grayscale tidak kurang dari 0 (nilai minimum)
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray) # Menyimpan nilai grayscale yang baru ke dalam setiap channel warna dari pixel tersebut
```

```python
#menampilkan hasil
nthrootpower(50, 100) # Memanggil fungsi nthpower dengan parameter
plt.imshow(img_nthrootpower) # Menampilkan citra hasil proses
plt.title("Nth Root Power") # Menambahkan judul
plt.show() # Menampilkan plot gambar
```
