```python
#import library
import numpy as np #untuk melakukan operasi matematika dan numerik, seperti pengolahan array dan matriks.
import imageio #library untuk membaca dan menulis berbagai format file gambar
import matplotlib.pyplot as plt #untuk membuat visualisasi grafik, diagram, dan plot data
```

```python
#membaca gambar
img = imageio.imread("mawar.jpg")
```

```python
#mendapatkan resolusi dan type dari gambar
img_height = img.shape[0] #Baris kode ini akan mengambil nilai dimensi height dari citra dan menyimpannya pada variabel
img_width = img.shape[1] #Baris kode ini akan mengambil nilai dimensi width dari citra dan menyimpannya pada variabel
img_channel = img.shape[2] # Baris kode ini akan mengambil nilai dimensi channel dari citra dan menyimpannya pada variabel
```

```python
#merubah gambar menjadi grayscale
img_grayscale = np.zeros(img.shape, dtype=np.uint8) # akan membuat citra baru yang berukuran sama dengan citra asli

for y in range(0, img_height): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi height dari citra
    for x in range(0, img_width): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi width dari citra 
        red = img[y][x][0] # akan mengambil nilai piksel untuk warna merah (red) pada koordinat (y, x) 
        green = img[y][x][1] # akan mengambil nilai piksel untuk warna hijau (green) pada koordinat (y, x) 
        blue = img[y][x][2] # akan mengambil nilai piksel untuk warna biru (blue) pada koordinat (y, x) 
        gray = (int(red) + int(green) + int(blue)) / 3 #akan menghitung nilai rata-rata dari ketiga nilai piksel warna (red, green, dan blue) pada koordinat (y, x) 
        img_grayscale[y][x] = (gray, gray, gray) #akan menempatkan nilai piksel yang sudah dikonversi menjadi citra grayscale pada koordinat (y, x)
        
plt.imshow(img_grayscale) # akan menampilkan citra hasil konversi ke citra grayscale yang telah disimpan pada variabel
plt.title("Grayscale") # akan memberikan judul pada plot 
plt.show() # akan menampilkan plot citra hasil konversi pada layar.
```

```python
#membuat variabel untuk menyimpan data gambar
hg = np.zeros((256)) #akan membuat sebuah array numpy satu dimensi dengan 256 elemen bernilai 0
```

```python
#mengisi setiap nilai dalam array hg dengan 0
for x in range(0, 256): # akan melakukan inisialisasi ulang (reset) nilai pada setiap elemen 
    hg[x] = 0
```

```python
#menampilkan histogram
for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        red = img[y][x][0] # nilai komponen merah (red) dari piksel pada baris ke-y, kolom ke-x
        green = img[y][x][1] # nilai komponen hijau (green) dari piksel pada baris ke-y, kolom ke-x
        blue = img[y][x][2] # nilai komponen biru (blue) dari piksel pada baris ke-y, kolom ke-x
        hgr[red] += 1 # tambahkan 1 ke elemen ke-red dari array hgr
        hgg[green] += 1 # tambahkan 1 ke elemen ke-green dari array hgg
        hgb[blue] += 1 # tambahkan 1 ke elemen ke-blue dari array hgb

bins = np.linspace(0, 256, 100) # inisialisasi range bin dan jumlah bin untuk histogram
plt.hist(hgr, bins, color="red", alpha=0.5) # plot histogram untuk komponen merah dengan warna merah transparan
plt.title("Histogram Red") # judul plot histogram untuk komponen merah
plt.show() # tampilkan plot histogram

plt.hist(hgg, bins, color="green", alpha=0.5) # plot histogram untuk komponen hijau dengan warna hijau transparan
plt.title("Histogram Green") # judul plot histogram untuk komponen hijau
plt.show() # tampilkan plot histogram

plt.hist(hgb, bins, color="blue", alpha=0.5) # plot histogram untuk komponen biru dengan warna biru transparan
plt.title("Histogram Blue") # judul plot histogram untuk komponen biru
plt.show()  # tampilkan plot histogram
```

```python
#menampilkan histogram kumulatif
hgk = np.zeros((256)) # inisialisasi array hgk dengan 256 elemen nol
c = np.zeros((256)) # inisialisasi array c dengan 256 elemen nol

for x in range(0, 256): # loop sebanyak 256 elemen pada array hgk dan c
    hgk[x] = 0 # set nilai elemen ke-x dari array hgk menjadi 0
    c[x] = 0 # set nilai elemen ke-x dari array c menjadi 0

for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        gray = img_grayscale[y][x][0] # nilai piksel grayscale pada baris ke-y, kolom ke-x
        hgk[gray] += 1 # tambahkan 1 ke elemen ke-gray dari array hgk
                
c[0] = hgk[0] # set nilai elemen ke-0 dari array c dengan nilai elemen ke-0 dari array hgk
for x in range(1, 256): # loop sebanyak 255 elemen pada array hgk dan c
     c[x] = c[x-1] + hgk[x] # set nilai elemen ke-x dari array c dengan penjumlahan elemen ke-(x-1) dan elemen ke-x dari array hgk

hmaxk = c[255] # simpan nilai elemen ke-255 dari array c ke dalam variabel hmaxk

for x in range(0, 256): # loop sebanyak 256 elemen pada array c
    c[x] = 190 * c[x] / hmaxk # set nilai elemen ke-x dari array c dengan perhitungan nilai elemen ke-x dikali 190 dan dibagi dengan hmaxk

plt.hist(c, bins, color="black", alpha=0.5) # plot histogram untuk array c dengan warna hitam transparan
plt.title("Histogram Grayscale Kumulatif") # judul plot histogram
plt.show() # tampilkan plot histogram
```

```python
#menampilkan histogram hequalisasi
hgh = np.zeros((256)) # inisialisasi array hgh dengan 256 elemen nol
h = np.zeros((256)) # inisialisasi array h dengan 256 elemen nol
c = np.zeros((256)) # inisialisasi array c dengan 256 elemen nol


for x in range(0, 256): # loop sebanyak 256 elemen pada array hgh, h, dan c
    hgh[x] = 0 # set nilai elemen ke-x dari array hgh menjadi 0
    h[x] = 0 # set nilai elemen ke-x dari array h menjadi 0
    c[x] = 0 # set nilai elemen ke-x dari array c menjadi 0

for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        gray = img_grayscale[y][x][0] # nilai piksel grayscale pada baris ke-y, kolom ke-x
        hgh[gray] += 1 # tambahkan 1 ke elemen ke-gray dari array hgh
                
h[0] = hgh[0] # set nilai elemen ke-0 dari array h dengan nilai elemen ke-0 dari array hgh
for x in range(1, 256): # loop sebanyak 255 elemen pada array h dan hgh
     h[x] = h[x-1] + hgh[x] # set nilai elemen ke-x dari array h dengan penjumlahan elemen ke-(x-1) dan elemen ke-x dari array hgh

for x in range(0, 256): # loop sebanyak 256 elemen pada array h
     h[x] = h[x] / img_height / img_width # set nilai elemen ke-x dari array h dengan perhitungan nilai elemen ke-x dibagi dengan hasil kali tinggi dan lebar gambar

for x in range(0, 256): # loop sebanyak 256 elemen pada array hgh
    hgh[x] = 0 # set nilai elemen ke-x dari array hgh menjadi 0
    
for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        gray = img_grayscale[y][x][0] # nilai piksel grayscale pada baris ke-y, kolom ke-x
        gray = h[gray] * 255 # ubah nilai piksel grayscale menjadi nilai baru dengan mengalikan nilai piksel grayscale sebelumnya dengan 255 dan kemudian membaginya dengan nilai elemen ke-nilai piksel grayscale baru dari array h
        hgh[int(gray)] += 1 # tambahkan 1 ke elemen ke-nilai piksel grayscale baru yang dibulatkan menjadi integer dari array hgh

c[0] = hgh[0] # set nilai elemen ke-0 dari array c dengan nilai elemen ke-0 dari array hgh
for x in range(1, 256): # loop sebanyak 255 elemen pada array hgh dan c
     c[x] = c[x-1] + hgh[x] # set nilai elemen ke-x dari array c dengan penjumlahan elemen ke-(x-1) dan elemen ke-x dari array hgh

hmaxk = c[255] # Mencari nilai elemen ke-255 dari array 'c' dan menyimpannya ke dalam variabel 'hmaxk'

for x in range(0, 256): #  Melakukan loop pada range 0 hingga 255
    c[x] = 190 * c[x] / hmaxk # Melakukan normalisasi pada elemen ke-x array c

plt.hist(c, bins, color="black", alpha=0.5) # Membuat histogram dengan data yang telah di-normalisasi dari array c
plt.title("Histogram Grayscale Hequalisasi") # Memberikan judul pada plot histogram
plt.show() # Menampilkan plot histogram pada layar
```
