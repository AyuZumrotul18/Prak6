```python
#import library
import numpy as np #untuk melakukan operasi matematika dan numerik, seperti pengolahan array dan matriks.
import imageio #library untuk membaca dan menulis berbagai format file gambar
import matplotlib.pyplot as plt #untuk membuat visualisasi grafik, diagram, dan plot data
```

```python
#membaca gambar
img = imageio.imread("daisy.jpg")
```

```python
#mendapatkan resolusi dan type dari gambar
img_height = img.shape[0] #Baris kode ini akan mengambil nilai dimensi height dari citra dan menyimpannya pada variabel
img_width = img.shape[1] #Baris kode ini akan mengambil nilai dimensi width dari citra dan menyimpannya pada variabe
img_channel = img.shape[2] # Baris kode ini akan mengambil nilai dimensi channel dari citra dan menyimpannya pada variabel
img_type = img.dtype #Baris kode ini akan mengambil tipe data dari citra dan menyimpannya pada variabel 
```

```python
#membuat variabel dengan resolusi dan tipe yang sama seperti gambar
img_flip_horizontal = np.zeros(img.shape, img_type) #Array ini akan digunakan untuk menyimpan citra hasil flip horizontal.
img_flip_vertical = np.zeros(img.shape, img_type) #Array ini akan digunakan untuk menyimpan citra hasil flip vertical.
```

```python
#membalik gambar secara horizontal
for y in range(0, img_height): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi height dari citra
    for x in range(0, img_width): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi width dari citra 
        for c in range(0, img_channel): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi width dari citra 
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] #akan menempatkan nilai pixel yang berada pada posisi (y, x, c) 
```

```python
#membalik gambar secara vertical
for y in range(0, img_height): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi height dari citra
    for x in range(0, img_width): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi width dari citra 
        for c in range(0, img_channel): #akan melakukan loop dari nilai 0 hingga yang merupakan nilai dimensi channel dari citra 
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] #akan menempatkan nilai pixel tersebut pada posisi (y, x, c) 
```

```python
#menampilkan hasil balik gambar
plt.imshow(img_flip_horizontal) # akan menampilkan citra hasil flip horizontal yang telah disimpan pada variabel
plt.title("Flip Horizontal") #akan memberikan judul pada plot citra hasil flip horizontal dengan teks "Flip Horizontal".
plt.show() # menampilkan plot citra hasil flip horizontal pada layar
plt.imshow(img_flip_vertical) # akan menampilkan citra hasil flip vertical yang telah disimpan pada variabel
plt.title("Flip Vertical") #akan memberikan judul pada plot citra hasil flip vertical dengan teks "Flip Vertical".
plt.show() #akan menampilkan plot citra hasil flip vertical pada layar.
```
